**II. Radar Chart**

```python
metric_cols = ["recall", "ndcg", "coverage", "novelty", "tail_fraction"]

# Define the correct approaches (temp or avg)
columns_st = [
    "Visual Unimodal - temp",
    "ST Unimodal - temp",
    "Visual + Textual (ST) CCA - temp",
]
columns_openai = [
    "Visual Unimodal - temp",
    "OpenAI Unimodal - temp",
    "Visual + Textual (OpenAI) CCA - temp",
]
columns_llama = [
    "Visual Unimodal - temp",
    "Llama3.0 Unimodal - temp",
    "Visual + Textual (Llama3.0) CCA - temp",
]

# For ba_df_llm_temp
# columns_st = [
#     "Visual Unimodal",
#     "ST Unimodal",
#     "Visual + Textual (ST) CCA",
# ]
# columns_openai = [
#     "Visual Unimodal",
#     "OpenAI Unimodal",
#     "Visual + Textual (OpenAI) CCA",
# ]
# columns_llama = [
#     "Visual Unimodal",
#     "Llama3.0 Unimodal",
#     "Visual + Textual (Llama3.0) CCA",
# ]
```

```python
import numpy as np
import matplotlib.pyplot as plt

custom_colors = ['teal', 'crimson', 'yellowgreen']

# Normalize data (min-max scaling to [0, 1])
def normalize(df):
    normalized_df = df.copy()
    for col in metric_cols:
        min_val = df[col].min()
        max_val = df[col].max()
        # Calculate the range
        rng = max_val - min_val
        if rng > 1e-9: # 1e-9
            # Perform min-max normalization
            normalized_df[col] = (df[col] - min_val) / rng
        else:
            # Fallback for columns where the range is zero (constant values)
            normalized_df[col] = 0.1
    return normalized_df

# Filter and get metrics for given columns
def get_metrics(df, columns):
    filtered_df = df[df['approach'].isin(columns)]
    metrics = filtered_df[metric_cols]
    labels = filtered_df['approach'].tolist()  # Get correct labels
    return metrics, labels

# Radar chart function
def radar_chart(ax, data, labels, title):
    angles = np.linspace(0, 2 * np.pi, len(metric_cols), endpoint=False).tolist()
    angles += angles[:1]  # Close the circle

    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)

    for idx, (row, label) in enumerate(zip(data.values, labels)):
        values = np.append(row, row[0])  # Close the loop
        color = custom_colors[idx % len(custom_colors)]
        ax.plot(angles, values, marker='p', label=label, color=color, linewidth=0.7)
        ax.fill(angles, values, alpha=0.25, color=color)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(metric_cols, fontsize=10, fontfamily='sans-serif')
    ax.set_yticklabels([])
    ax.set_title(title, fontsize=16, fontfamily='sans-serif', pad=35)
    ax.legend(loc="upper center", bbox_to_anchor=(0.5, -0.05), fontsize=12)

# Normalize the entire dataframe (*** temp or avg)
ba_df_llm_normalized = normalize(ba_df_llm)

# Create radar charts
fig, axs = plt.subplots(1, 3, figsize=(18, 6), subplot_kw=dict(polar=True))

# ST Models Radar Chart
st_data, st_labels = get_metrics(ba_df_llm_normalized, columns_st)
radar_chart(axs[0], st_data, st_labels, "ST Models")

# OpenAI Models Radar Chart
openai_data, openai_labels = get_metrics(ba_df_llm_normalized, columns_openai)
radar_chart(axs[1], openai_data, openai_labels, "OpenAI Models")

# Llama Models Radar Chart
llama_data, llama_labels = get_metrics(ba_df_llm_normalized, columns_llama)
radar_chart(axs[2], llama_data, llama_labels, "Llama3.0 Models")

plt.tight_layout()
plt.savefig("radar_charts.pdf", format='pdf', bbox_inches='tight')
plt.show()
```

**II. Bar Chart Combined**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Metrics to plot
metrics = ["Recal (temp)", "NDCG (temp)"]

# Plot setup
fig, axes = plt.subplots(1, 2, figsize=(18, 9))
axes = axes.flatten()
bar_width = 0.2
spacing = 2.0
x = range(len(df_ccaFused_ret["N Retrieval"]))

# Define datasets and labels
datasets = [
    ("Fused (CCA) Retrieval", df_ccaFused_ret),
    ("Fused (CCA) Rec. Manual", df_ccaFused_recMan),
    ("Fused (CCA) Rec. LLM", df_ccaFused_recLLM),
    ("Textual Retrieval", df_data_st_ret),
    ("Textual Rec. Manual", df_st_recMan),
    ("Textual Rec. LLM", df_st_recLLM),
    ("Visual Retrieval", df_data_vis_ret),
    ("Visual Rec. Manual", df_vis_recMan),
    ("Visual Rec. LLM", df_vis_recLLM)
]

# Colors and markers for trend lines
colors = plt.cm.tab10.colors

# Plot bars with rounded edges and trend lines
for i, metric in enumerate(metrics):
    for j, (label, df) in enumerate(datasets):
        bar_positions = [p * spacing + bar_width * j for p in x]
        bars = axes[i].bar(
            bar_positions, df[metric], bar_width,
            label=label, color=colors[j % len(colors)]
        )

        # Apply border radius (rounded corners)
        for bar in bars:
            bar.set_edgecolor('black')
            bar.set_linewidth(1.5)
            bar.set_alpha(0.9)

        # Trend lines with reduced opacity
        midpoints = [pos + bar_width / 2 for pos in bar_positions]
        axes[i].plot(midpoints, df[metric], marker='o', color=colors[j % len(colors)], alpha=0.5)

    # Adjustments for better readability
    axes[i].set_xticks([p * spacing + bar_width * (len(datasets) // 2) for p in x])
    axes[i].set_xticklabels(df_ccaFused_ret["N Retrieval"], fontsize=14)
    axes[i].set_xlabel("Retrieval Depth (Top-N)", fontsize=16)
    axes[i].set_ylabel(metric, fontsize=16)
    # axes[i].set_title(f"Comparison of {metric} across all models", fontsize=16)

# Create a single legend below all plots
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(
    handles, labels,
    loc='lower center',
    fontsize=12,
    ncol=5,
    bbox_to_anchor=(0.5, -0.05)
)

# Adjust spacing between charts
plt.subplots_adjust(wspace=0.4)

plt.tight_layout(rect=[0, 0.05, 1, 1])
plt.savefig("ablation.pdf", format='pdf', bbox_inches='tight')
plt.show()
```
